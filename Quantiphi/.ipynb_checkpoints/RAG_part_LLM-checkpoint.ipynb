```python
# Download the PDF and select chapters
pdf_path = "book/ConceptsofBiology-WEB.pdf"

def extract_text_from_pdf(pdf_path, chapter_numbers):
    with fitz.open(pdf_path) as pdf_document:
        text = ""
        print(pdf_document.page_count)
        print(pdf_document)
        for page_number in range(chapter_numbers):
            page = pdf_document.load_page(page_number)
            text += page.get_text()
    return text

# No in-memory DB used
def create_index(text):
    paragraphs = text.split("\n")  # Split text into paragraphs
    index = {}
    start = 0
    for i, para in enumerate(paragraphs):
        end = start + len(para)  # Calculate end index of each paragraph
        index[f"Paragraph {i+1}"] = [start, end]
        start = end + 1  # Update start index for the next paragraph
    return index
```

```python
selected_chapters = [0, 1]  # Adjust chapter numbers accordingly
text = extract_text_from_pdf(pdf_path, selected_chapters)
```
